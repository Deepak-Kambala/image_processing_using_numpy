```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt


image = Image.open('wild.gif').convert('RGB')


img_array = np.array(image)


negative_array = 255 - img_array


negative_image = Image.fromarray(negative_array)


plt.figure(figsize=(10, 7))

plt.subplot(1, 2, 1)
plt.title('Uploaded Image')
plt.imshow(img_array)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Changed Image')
plt.imshow(negative_array)
plt.axis('off')

plt.tight_layout()
plt.show()

negative_image.save('output.jpg')
```
